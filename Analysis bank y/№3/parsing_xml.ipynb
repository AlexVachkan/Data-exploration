```python
from bs4 import BeautifulSoup
from datetime import datetime
import pandas as pd
```

```python
def parser(path):
    '''
    Принимает путь к файлу xml
    Выдает датафрейм
    '''
    
    fd = open(path, 'r') 
    xml_file = fd.read() 
    soup = BeautifulSoup(xml_file, 'lxml') 

    for i in soup.findAll("group_id"):
        group_id = i.text 

    for i in soup.findAll("create_datetime"): 
        create_datetime =  pd.to_datetime(i.text, 
                           format='%d%m%Y%H%M%S', 
                           errors='ignore').date().strftime("%d-%m-%Y") 

    open_date=[]
    relationship=[]
    final_pmt_date=[]
    pmt_string_84m=[]
    status=[]
    outstanding=[]
    inf_confirm_date=[]
    credit_limit=[]
    delq_balance=[]
    is_own=[]
    credit_cost_rate=[]
    uuid=[]

    for i in soup.findAll("loan"):

        try:
            open_date.append(i.find("open_date").text)
        except: 
            open_date.append('Нет данных')

        try:
            relationship.append(i.find("relationship").text)
        except:   
            relationship.append('Нет данных')

        try:
            final_pmt_date.append(i.find("final_pmt_date").text)
        except: 
            final_pmt_date.append('Нет данных')

        try:
            pmt_string_84m.append(i.find("pmt_string_84m").text)
        except: 
            pmt_string_84m.append('Нет данных')

        try:
            status.append(i.find("status").text)
        except: 
            status.append('Нет данных')

        try:
            outstanding.append(i.find("outstanding").text)
        except:  
            outstanding.append('Нет данных')

        try:
            inf_confirm_date.append(i.find("inf_confirm_date").text)
        except:
            inf_confirm_date.append('Нет данных')

        try:
            credit_limit.append(i.find("credit_limit").text)
        except:
            credit_limit.append('Нет данных')

        try:
            delq_balance.append(i.find("delq_balance").text)
        except: 
            delq_balance.append('Нет данных')

        try:
            is_own.append(i.find("is_own").text)
        except:
            is_own.append('Нет данных')

        try:
            credit_cost_rate.append(i.find("credit_cost_rate").text)
        except:
            credit_cost_rate.append('Нет данных')

        try:    
            uuid.append(i.find("uuid").text)
        except: 
            uuid.append('Нет данных')

    d={
       'group_id' :group_id, 
       'create_datetime': create_datetime, 
       'open_date': open_date,
        'relationship': relationship,
        'final_pmt_date': final_pmt_date,
        'pmt_string_84m': pmt_string_84m,
        'status': status,
        'outstanding': outstanding,
        'inf_confirm_date': inf_confirm_date,
        'credit_limit': credit_limit,
        'delq_balance': delq_balance,
        'is_own': is_own,
        'credit_cost_rate': credit_cost_rate,
        'uuid': uuid
        }

    df = pd.DataFrame(data=d) 
    return df
```

```python
def to_csv(frames):
    '''
    Принимает список таблиц
    Выдает csv файл
    '''
    res = pd.concat(frames, ignore_index=True)

    res.to_csv('credit_report.csv', sep=',', encoding='utf-8', index=False)
```

```python
frames = [parser(path), parser(path_2), parser(path_3)]

to_csv(frames)
```

```python
def to_excel(frames):
    '''
    Принимает список таблиц
    Выдает csv файл
    '''
    res = pd.concat(frames, ignore_index=True)

    res.to_excel('credit_report.xlsx', index=False)
```
