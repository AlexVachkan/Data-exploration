```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
```

```python
page = 100

id_ = []
name = []
brand = []
volume = []
price = []
best_loyalty_price = []

for i in range(0, page+1):
    
    URL = f'https://api.rivegauche.ru/rg/v1/newRG/products/search?fields=FULL&currentPage=0&pageSize=24&categoryCode=Perfumery_Men&tag=4108894044469370'
    data = requests.get(URL)
    print(data.status_code)
    soup = BeautifulSoup(data.text, 'html.parser') # получаем суп страницы 
    
    data = json.loads(str(soup)) # преобразовываем файл суп в файл json 
    
    for j in range(len(data['results'])):
    
        id_.append(data['results'][j]['code']) # создаем cписки
        name.append(data['results'][j]['name']) # создаем cписки
        brand.append(data['results'][j]['manufacturer']) # создаем cписки
        
        if data['results'][j]['classifications'][0]['features'][3]['featureValues'][0]['value'] != "": # создаем cписки
            volume.append(data['results'][j]['classifications'][0]['features'][3]['featureValues'][0]['value'])
        else:
            volume.append('нет объема')
            
        price.append(data['results'][j]['prices'][2]['value']) # создаем cписки
        
        try:
            best_loyalty_price.append(data['results'][j]['price']['value']) # создаем cписки
            
        except:
            best_loyalty_price.append('нет скидки')
```

```python
df=pd.DataFrame() # создаем дф

df['id'] = id_ # добавляем столбцы в дф
df['name'] = name # добавляем столбцы в дф
df['brand'] = brand # добавляем столбцы в дф
df['volume'] = volume # добавляем столбцы в дф
df['price'] = price # добавляем столбцы в дф
df['best_loyalty_price'] = best_loyalty_price # добавляем столбцы в дф
```

```python
df.loc[:, "shop"] = 'Рив Гош' # добавляем столбцес магазином
```
