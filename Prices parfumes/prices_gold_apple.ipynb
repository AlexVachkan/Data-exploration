```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
```

```python
page = 100

id_ = []
name = []
brand = []
volume = []
price = []
best_loyalty_price = []

for i in range(0, page+1):
    
    URL = f'https://goldapple.ru/web_scripts/discover/category/products?cat=59&page={i}'
    data = requests.get(URL)
    print(data.status_code)
    soup = BeautifulSoup(data.text, 'html.parser') # получаем суп страницы 
    
    data = json.loads(str(soup)) # преобразовываем файл суп в файл json 
    
    for j in range(len(data['products'])):
    
        id_.append(data['products'][j]['id']) # создаем cписки
        name.append(data['products'][j]['name']) # создаем cписки
        brand.append(data['products'][j]['brand']) # создаем cписки
        
        if data['products'][j]['volume'] != "": # создаем cписки
            volume.append(data['products'][j]['volume'])
        else:
            volume.append('нет объема')
            
        price.append(data['products'][j]['price']) # создаем cписки
        
        try:
            best_loyalty_price.append(data['products'][j]['best_loyalty_price']) # создаем cписки
            
        except:
            best_loyalty_price.append('нет скидки')

    
```

```python
df=pd.DataFrame() # создаем дф

df['id'] = id_ # добавляем столбцы в дф
df['name'] = name # добавляем столбцы в дф
df['brand'] = brand # добавляем столбцы в дф
df['volume'] = volume # добавляем столбцы в дф
df['price'] = price # добавляем столбцы в дф
df['best_loyalty_price'] = best_loyalty_price # добавляем столбцы в дф
```

```python
df.loc[:, "shop"] = 'Золотое яблоко' # добавляем столбцес магазином
```
